```python
import pandas as pd
import statistics as st
import numpy as np
```

```python
df = pd.read_csv(r'verwerking/staal3/staal3_1.csv');
print(df)
coorX = df.X
coorY = df.Y
distMin = 3;
distMax = 30;
pixelSize=39;
k=0;
agg=[];
agg.clear();
dist_fin=[];
dist_fin.clear();
distance=[];
```

```python
for i in range(0, len(coorX)):
    distance.clear();
    k=0;
    for j in range(0, len(coorX)):
        dist = (((coorX[i]-coorX[j])**2+(coorY[i]-coorY[j])**2)**0.5)
        if (distMin<dist) and (dist<distMax) and (dist != 0) : distance.append(dist)
        if dist<distMin and (dist != 0):k=k+1;
        print (distance);
        print("k=", k);
    agg.append(k);
    if len(distance)>0: dist_fin.append(st.mean(distance)); print('mean', i, "=",st.mean(distance));
 
```

```python
print(len(dist_fin))
print(dist_fin)
print(agg)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=dist_fin, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.xlabel('Diameter, um')
plt.ylabel('Frequency, %')
plt.title('My Very Own Histogram')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
